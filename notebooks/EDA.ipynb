```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.max_rows = 4000
pd.options.display.max_colwidth = 200

%matplotlib inline
```

```python
def readData(path):
    
    with open(path, "r", encoding="utf-8") as file:
        data = json.load(file)

    TEXT = []
    ENTITY_NAME = []
    ENTITY_LABEL = []
#     IDS = []
    for i, dataItem in enumerate(data):
        entitiesDict = {}
        entitiesDict["entities"] = []
        for slotItem in dataItem["slots"]:
            TEXT.append(dataItem["text"])
            ENTITY_NAME.append(slotItem["slotValue"])
            ENTITY_LABEL.append(slotItem["slotName"])
            
    return pd.DataFrame({"text": TEXT, "name": ENTITY_NAME, "label": ENTITY_LABEL})
```

```python
def checkDuplicateEntities(path):
    
    with open(path, "r", encoding="utf-8") as file:
        data = json.load(file)

    cnt=0
    TEXT = []
    ENTITY_NAME = []
    ENTITY_LABEL = []
    IDS = []
#     IDS = []
    for i, dataItem in enumerate(data):
        entitiesDict = {}
        entitiesDict["entities"] = []
        TEXT.append(dataItem["text"])
        if dataItem["id"] in IDS:
            IDS.append(dataItem["id"])
            print(dataItem["id"], dataItem["text"])
            cnt+=1
        else:
            IDS.append(dataItem["id"])
            
    return pd.DataFrame({"text": TEXT, "id": IDS})
    
    print(cnt)
```
